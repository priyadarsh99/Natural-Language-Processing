##Tokenization

###Using Regex tokenizer

```python
reg_token = regexp_tokenize(raw,"[\w']+")
#using regular expression tokenizer 
#to convert the raw text into tokens
#as it doesn't separate the apostrophy symbol from the word
#hence better tokenization
```

```python
reg_token
```

```python
len(reg_token)
```

###Removing Stop Words

```python
nltk.download('stopwords')
#downloading the stopwords package
```

```python
stop_word = set(stopwords.words('english'))
#storing all the stopwords in the set for faster execution
```

```python
stop_word
```

```python
type(stop_word)
```

```python
#removing all the stopwords from the reg_token
filtered_token =[]
for i in reg_token:
  x = i.lower()
  if not( x in stop_word):
    filtered_token.append(x)
```

```python
len(reg_token),len(filtered_token)
```

```python
filtered_token
```

##Stemming

```python
#function to perform the stemming process using Snowball Stemmer
#Snowball Stemmer as it is more effective and efficient than the Porter and Lancaster Stemmer
def Stemming():
  snowball = SnowballStemmer(language='english')
  stemmed_list=[]
  for i in filtered_token:
    x = snowball.stem(i)
    print(i,'------->',x)#showing the original token and stemmed token
    stemmed_list.append(x)
  return(stemmed_list)
```

```python
print(Stemming())
```

#####Lemmatization is more effective than stemming as the tokens stored in it is more meaningful

##Lemmatization

```python
nltk.download('wordnet')#download the wordnet for lemmatization
```

```python
lemmatizer = WordNetLemmatizer() #creating object of wordNetlemmatizer
```

```python
#function to perform lemmatization on filtered_token
def lemmatization():
  lemmatized_list = []
  for i in filtered_token:
    x = lemmatizer.lemmatize(i)
    print(i,'------>',x)#showing original token and lemmatized token
    lemmatized_list.append(x)
  return(lemmatized_list)
```

```python
print(lemmatization())
```
